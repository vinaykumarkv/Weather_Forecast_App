```python
import requests
api_key=""
p="Bengaluru"
op="temperature_2m"
d="2"
```

```python
location = requests.get(f"https://geocode.maps.co/search?q={p}&api_key={api_key}")
longitude = location.json()[0]["lon"]
print(longitude)
```

```python
latitude = location.json()[0]["lat"]
print(latitude)
```

```python
r = requests.get(f"https://api.open-meteo.com/v1/forecast?latitude={latitude}&longitude={longitude}&hourly={op}&forecast_days={d}")
dates = r.json()["hourly"]["time"]
```

```python
t = r.json()["hourly"]["temperature_2m"]
```

```python
print(dates)
print(t)
```
